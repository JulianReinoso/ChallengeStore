### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
TotalTienda1 = tienda['Precio'].sum()
print(f"Facturacion total Tienda 1: {TotalTienda1}")

TotalTienda2 = tienda2['Precio'].sum()
print(f"Facturacion total Tienda 2: {TotalTienda2}")

TotalTienda3 = tienda3['Precio'].sum()
print(f"Facturacion total Tienda 3: {TotalTienda3}")

TotalTienda4 = tienda4['Precio'].sum()
print(f"Facturacion total Tienda 4: {TotalTienda4}")
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
print(f"--- Ventas por Categoria Tienda 1 --- \n {ventas_categoria_tienda1}")

ventas_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
print(f"\n--- Ventas por Categoria Tienda 2 --- \n {ventas_categoria_tienda2}")

ventas_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
print(f"\n--- Ventas por Categoria Tienda 3 --- \n {ventas_categoria_tienda3}")

ventas_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
print(f"\n--- Ventas por Categoria Tienda 4 --- \n {ventas_categoria_tienda4}")
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
tienda1_calificacion_prom = tienda['Calificación'].mean()
print(f"Calificacion promedio Tienda 1: {tienda1_calificacion_prom:.2f}\n")

tienda2_calificacion_prom = tienda2['Calificación'].mean()
print(f"Calificacion promedio Tienda 2: {tienda2_calificacion_prom:.2f}\n")

tienda3_calificacion_prom = tienda3['Calificación'].mean()
print(f"Calificacion promedio Tienda 3: {tienda3_calificacion_prom:.2f}\n")

tienda4_calificacion_prom = tienda4['Calificación'].mean()
print(f"Calificacion promedio Tienda 4: {tienda4_calificacion_prom:.2f}")
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df_actual in enumerate(lista_tiendas):
  numero_tienda = i + 1
  conteo = df_actual['Producto'].value_counts()
  top3 = conteo.head(3)
  bottom3 = conteo.tail(3)

  print(f"\n--- Tienda {numero_tienda} ---")
  print("Productos más vendidos:")
  print(top3)
  print("\nProductos menos vendidos:")
  print(bottom3)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Agrupamos las tiendas en una lista
lista_tiendas = [tienda, tienda2, tienda3, tienda4]

# 2. Preparamos el "lienzo" con 4 subgráficos (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten() # Esto nos permite acceder a los gráficos como 0, 1, 2, 3

# 3. Bucle para generar cada gráfico
for i, df_actual in enumerate(lista_tiendas):

    # Calcular conteos
    conteo = df_actual['Producto'].value_counts()

    # Obtener extremos
    top_3 = conteo.head(3)
    bottom_3 = conteo.tail(3)

    # Unirlos en un solo objeto para graficar
    # Usamos .iloc[::-1] para invertir el orden y que el #1 salga arriba visualmente
    combined = pd.concat([top_3, bottom_3]).iloc[::-1]

    # Definir colores: Los primeros 3 (que son los bottom) en rojo, los últimos 3 (top) en azul
    colores = ['salmon']*3 + ['skyblue']*3

    # Graficar en el eje correspondiente (axes[i])
    combined.plot(kind='barh', ax=axes[i], color=colores, edgecolor='grey')

    # Títulos y etiquetas
    axes[i].set_title(f'Tienda {i+1}: Top 3 (Azul) vs Bottom 3 (Rojo)')
    axes[i].set_xlabel('Cantidad de Ventas')

# Ajustar el espaciado para que no se encimen los textos
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
lista_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df_actual in enumerate(lista_tiendas):
  numero_tienda = i + 1
  promedio = df_actual['Costo de envío'].mean()
  print(f"\n--- Tienda {numero_tienda} ---")
  print(f"Costo de envio promedio: {promedio:.2f}")
```

# 6. Graficos

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Configuración estética opcional para que los gráficos se vean más modernos
sns.set_theme(style="whitegrid")
```

```python
# --- PREPARACIÓN DE DATOS ---

# 1. Datos para Ingresos (Gráfico de Barras)
tiendas_labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos_totales = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]

# 2. Datos para Categorías (Barras Apiladas)
# Creamos un DataFrame combinado solo para esto
df_cat_temp = pd.DataFrame({
    'Tienda 1': tienda['Categoría del Producto'].value_counts(),
    'Tienda 2': tienda2['Categoría del Producto'].value_counts(),
    'Tienda 3': tienda3['Categoría del Producto'].value_counts(),
    'Tienda 4': tienda4['Categoría del Producto'].value_counts()
}).transpose() # Transponemos para que las tiendas sean las filas

# 3. Datos para Calificaciones (Boxplot)
# Combinamos las calificaciones en una lista para Seaborn
datos_calificaciones = [
    tienda['Calificación'],
    tienda2['Calificación'],
    tienda3['Calificación'],
    tienda4['Calificación']
]

# --- CREACIÓN DE LOS GRÁFICOS ---

# Creamos una figura grande con 2 filas y 2 columnas (usaremos 3 espacios)
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Panel de Análisis de Rendimiento de Tiendas', fontsize=20, fontweight='bold')

# GRÁFICO 1: Comparación de Ingresos Totales (Barras Verticales)
# Usamos axes[0, 0] (fila 0, columna 0)
sns.barplot(x=tiendas_labels, y=ingresos_totales, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Facturación Total por Tienda', fontsize=14)
axes[0, 0].set_ylabel('Ingresos ($)', fontsize=12)
# Formato para que el eje Y no muestre notación científica (opcional)
axes[0, 0].ticklabel_format(style='plain', axis='y')


# GRÁFICO 2: Distribución de Categorías (Barras Apiladas)
# Usamos axes[0, 1] (fila 0, columna 1)
# Pandas tiene un método plot integrado muy útil para barras apiladas
df_cat_temp.plot(kind='bar', stacked=True, ax=axes[0, 1], colormap='tab10')
axes[0, 1].set_title('Composición de Ventas por Categoría', fontsize=14)
axes[0, 1].set_ylabel('Cantidad de Ventas', fontsize=12)
axes[0, 1].legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')


# GRÁFICO 3: Distribución de Calificaciones (Boxplot)
# Usamos axes[1, 0] (fila 1, columna 0) y ocupamos todo el ancho inferior
plt.subplot(2, 1, 2) # Redefinimos el espacio inferior para que sea un solo gráfico ancho
sns.boxplot(data=datos_calificaciones, palette='Set2')
plt.xticks(ticks=[0, 1, 2, 3], labels=tiendas_labels)
plt.title('Distribución de Calificaciones de Clientes (1-5 estrellas)', fontsize=14)
plt.ylabel('Calificación', fontsize=12)
plt.ylim(0.5, 5.5) # Asegurar que se vea el rango completo de 1 a 5

# Ajuste final para que no se superpongan los textos
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Dejamos espacio para el título principal
plt.show()
```
